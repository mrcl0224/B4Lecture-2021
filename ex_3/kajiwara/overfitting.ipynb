```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import mean_absolute_error,mean_squared_error

from linear_regression import LinearRegression
```

```python
df1 = pd.read_csv('../data1.csv')
df2 = pd.read_csv('../data2.csv')
df3 = pd.read_csv('../data3.csv')
```

```python
# data1
x_data1 = df1['x1'].values.reshape(-1, 1)
y_data1 = df1['x2'].values
```

```python
digrees = [1, 3, 5, 10, 15]
models = []
linex = np.linspace(-5, 10)
```

```python
fig, ax = plt.subplots()
for d in digrees:
    model = LinearRegression(digree=d)
    model.fit(x_data1, y_data1)
    models.append(model)

    y_ = model.predict(linex.reshape(-1, 1))
    
    ax.plot(linex, y_, label='digree={}'.format(d))

ax.set_xlim([-6, 10])
ax.set_ylim([-10, 20])
ax.scatter(x_data1, y_data1, label='obserbed')
ax.legend(loc=4)
ax.set(title="", xlabel="", ylabel="")
plt.savefig('var_dig_1.png')
```

```python
for i, model in enumerate(models):
    pred = model.predict(x_data1)
    mse = mean_squared_error(y_data1, pred)
    print('digree = {}: {}'.format(digrees[i], mse))
```
