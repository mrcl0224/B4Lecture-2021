```python
from linear_regression import LinearRegression

model = LinearRegression(digree=2)
```

```python
X1 = df3['x1'].values
X2 = df3['x2'].values
y = df3['x3'].values
```

```python
X = np.array([X1, X2]).T
```

```python
model.fit(X, y)
```

```python
model.coef
```

```python
w = model.coef
# y_ = w[0]*X[:,0]+w[1]*X[:,1]+w[2]
linex1 = np.linspace(-4, 4)
linex2 = np.linspace(0, 10)

ax_1, ax_2 = np.meshgrid(linex1, linex2)
y_ = w[4]*ax_2 + w[5]*ax_1 + w[1]*ax_2**2 + w[2]*ax_1*ax_2 + w[3]*ax_1**2 + w[0]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, label='Observed')
ax.plot_wireframe(ax_1, ax_2, y_, color='indianred', label='Prediction')
```

```python
v_count = X.shape[1]
```

```python
phi = np.empty((len(X), 1+1+v_count))
for v in range(v_count):
    _X = X[:, v]
    _phi = np.zeros((len(_X), 1+1))
    for i in range(1+1):
        p = _X**i
        _phi[:, i] = p.ravel()
    phi[:, v*2:v*2+2]
```
